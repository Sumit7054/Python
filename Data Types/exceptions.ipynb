This part shows how to tackle exceptions in python.

Try and Catch is generally used to handle errors gracefully. Try is used to test if the code part below runs successfully with no runtime errors.



Except statement is used to catch the exception or errors. A logic below this statement is run if any error occured in try statement.

```python
try:
    value:bool = 3/0
except Exception as e:
    print(f"There was an error: {e}")
```

In above code, you could see that 3 can not be divided by 0 and therefore divide by zero error is returned.

The code prints zero division error. And the except is not raised. The code runs error free.
