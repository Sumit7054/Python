This part shows how to tackle exceptions in python.

Try and Catch is generally used to handle errors gracefully. Try is used to test if the code part below runs successfully with no runtime errors.



Except statement is used to catch the exception or errors. A logic below this statement is run if any error occured in try statement.

```python
try:
    value:bool = 3/0
except Exception as e:
    print(f"There was an error: {e}")
```

In above code, you could see that 3 can not be divided by 0 and therefore divide by zero error is returned.

The code prints zero division error. And the except is not raised. The code runs error free.

We can also raise an error if we want to stop the code execution on error, like so.

```python
try:
    value:bool = 3/0
    print("This line will not be executed if there is an error above.")
except Exception as e:
    print(f"There was an error: {e} and therefore raising it again.")
##    raise
```

You can  also try else statement if the execution of code is successful without error.

```python
try:
    value = 3/1
except ZeroDivisionError as zde:
    print(f"Zero Division Error occurred: {zde}")
else:
    print(f"The code executed successfully and the value is {value}.")
    
```

Same code will not enter in else if there is a zero divion error!

```python
try:
    value = 3/0
except ZeroDivisionError as zde:
    print(f"Zero Division Error occurred: {zde}. Code will not enter in else block.")
else:
    print(f"The code executed successfully and the value is {value}.")
```
