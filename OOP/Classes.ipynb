In this section, we will discuss about class concepts in Python.


Class is a form of template, design or architecture which can help us make products of that template.
An instance of class is called object.

```python
class Teacher:
    pass
```

In above case, we start a class called Teacher and we have no attributes or actions in it. We can create objects from this class template.

```python
T1 = Teacher()
```

This class template does not have any attributes or possibility of actions.

We can also create public attributes as well methods in it.

```python
class Teacher:
    def __init__(self, name, subject, qualification, age, level):
        self.name = name
        self.subject = subject
        self.qualification = qualification
        self.age = age
        self.level = level
    def introduce(self):
        return f"Hello, my name is {self.name}. I teach {self.subject} and I have a {self.qualification} degree."
    
    def is_eligible_for_retirement(self):
        if self.level == "primary" and self.age >= 60:
            return True
        elif self.level == "secondary" and self.age >= 62:
            return True
        elif self.level == "university" and self.age >= 65:
            return True
        else:
            return False
    
    def promote(self):
        if self.level == "primary":
            self.level = "secondary"
        elif self.level == "secondary":
            self.level = "university"
        else:
            return "Already at highest level"
        return f"Promoted to {self.level} level"
    
    def change_subject(self, new_subject):
        self.subject = new_subject
        return f"Subject changed to {self.subject}"
    def birthday(self):
        self.age += 1
        return f"Happy Birthday {self.name}! You are now {self.age} years old."    
```

This class initiation is done with __init__ dunder method. It has certain attributes which are relevant for the teacher. There are some methods also which can help access some of the attributes as well help change some of them.

```python
T1 = Teacher("John Doe", "Mathematics", "PhD", 45, "secondary")
```

The above line creates the object T1 from Teacher template. 

You can actually get this teacher to introduce itself.

```python
T1.introduce()
```

```python
print("The teacher name is: ", T1.name)
```

You can also create a class with private attributes. Those attributes can only be accessed by get method and can also be set by set methods only.

```python
class Car:
    def __init__(self, make, model, year, mileage):
        self.__make = make
        self.__model = model
        self.__year = year
        self.__mileage = mileage

    def get_make(self):
        return self.__make

    def set_make(self, make):
        self.__make = make

    def get_model(self):
        return self.__model

    def set_model(self, model):
        self.__model = model

    def get_year(self):
        return self.__year

    def set_year(self, year):
        self.__year = year

    def get_mileage(self):
        return self.__mileage

    def set_mileage(self, mileage):
        self.__mileage = mileage
```

The private attributes are created with __ before the attribute name.

```python
carA = Car("Toyota", "Camry", 2020, 15000)
```

You can access these carA attributes through get methods only. Direct access is forbidden.

```python
carA.get_make()
```

carA.__make is likely to give you the AttributeError saying this attribute does not exist.


The class can also have class methods which does not require object creation.

```python
class Plane:
    floors = 2
    @classmethod
    def plane_info(cls):
        return f"This plane has {cls.floors} floors."
    def __init__(self, model, capacity):
        self.model = model
        self.capacity = capacity
        
    def plane_details(self):
        return f"Model: {self.model}, Capacity: {self.capacity} passengers."
    
    def set_capacity(self, capacity):
        self.capacity = capacity
        return f"Capacity updated to {self.capacity} passengers."
    
    def get_capacity(self):
        return self.capacity
    def get_model(self):
        return self.model
```

You can create object from this class.

```python
planeA = Plane("Boeing 747", 416)
```

```python
planeA.plane_info()
```

```python
planeA.get_model()
```

```python
planeA.floors
```

```python
planeA.set_capacity(500)
```

The above set of code would have been able to explain public and private attributes, class method and class attributes.

# Lets have a look at inheritance.

```python
class Animal:
    def __init__(self, species, age):
        self.species = species
        self.age = age

    def make_sound(self):
        print("Some generic animal sound - poo!")

    def birthday(self):
        self.age += 1
        return f"Happy Birthday! The {self.species} is now {self.age} years old."
```

```python
class Buffalo(Animal):
    def __init__(self, age, horn_length):
        super().__init__("Buffalo", age)
        self.horn_length = horn_length
        
    def make_sound(self):
        return "Moo"
```

```python
buffaloA = Buffalo(5, 30)
```

```python
buffaloA.make_sound()
```
